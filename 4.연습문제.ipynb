양수음수분류기
다양한 개수의 정수를 입력받아, 각 수가 양수인지 음수인지 구분하여 각각의 개수를 출력하는 코드를 작성하시오.
- 입력으로부터 정수리스트로 변환하는 것은 한 줄 코드로 작성
- 입력 받은 값 중 0은 무시
- quit 를 입력하면 프로그램을 중지합니다.
---
출력예시

정수들을 입력: 입력한 숫자가 없습니다 양수: 0개, 음수: 0개, 합계: 0
정수들을 입력: 12 -56 25 98 -54 0 77 0 양수: 4개, 음수: 2개, 합계: 102

````python
while True: 
    user_input = input('정수의 목록을 공백으로 구분해서 입력하세요. 종료: "q"')
    #종료 명령이면 break
    if user_input == 'q' :
        break
    str_int = user_input.split()
    numbers = [] #초기값
        
    #int(user_input)
    for num in str_int:
        # if num == '0' :
        #     continue
        if num != '0':
            numbers.append(int(num))
    
    #입력받은 정수 리스트
    print(numbers)
    #입력한 숫자가 없습니다 체크
    if numbers == None: 
        print('입력한 숫자가 없습니다')
    #양수/음수 판정
    plus_numbers = []
    minus_numbers = []
    for num in numbers:
        if num > 0 :
            plus_numbers.append(num)
        else : 
            minus_numbers.append(num)
    pls_cnt = len(plus_numbers)
    minus_cnt = len(minus_numbers)
    total = sum(numbers)
    print(f'양수 {pls_cnt}개 음수 {minus_cnt}개 총 합게 : {total}')
    print()
````
