```python
import requests
btc = requests.get("http://api.bithumb.com/public/ticker/BTC").json()['data']
```

```python
시가 = int(btc['opening_price'])
최고가 = int(btc['max_price'])
최저가 = int(btc['min_price'])
변동폭 = 최고가 - int(btc['min_price'])
변동폭
```

```python
if 시가+변동폭 > 최고가 : 
    print('상승장')
else : 
    print('하락장')
```
