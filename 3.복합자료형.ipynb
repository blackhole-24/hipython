# 리스트와 for문의 활용

```python
students = [{'num':'1','name':'kim', 'kor':80,'math':90,'eng':100,},
            {'num':'2','name':'park', 'kor':90,'math':60,'eng':80,},
            {'num':'3','name':'hong', 'kor':70,'math':70,'eng':80,},
            {'num':'4','name':'hong1', 'kor':70,'math':70,'eng':80,},
            {'num':'5','name':'hong2', 'kor':70,'math':70,'eng':80,},
           ]
```

```python
print(students[0]) #for 문으로 간단히!!
print(students[1])
print(students[2])
print(students[3])
print(students[4])
```

```python
#인덱싱하기
students[2]['kor']
```

```python
for stu in students:
    stu['total'] = stu['kor']+stu['math']+stu['eng']
    print(stu['total'])
    #print('\n')
```

```python
students
```

```python
# 평균 구하기
for stu in students:
    stu['avg'] = round(stu['total']/3,2)
```

```python
for stu in students:
    print(stu)
```
