```python
import numpy as np
from scipy import stats

# ==========================================
# 1ë‹¨ê³„: ë°ì´í„° ìˆ˜ì§‘ (ê°€ìƒ ë°ì´í„° ìƒì„±)
# ==========================================
np.random.seed(42)  # ê²°ê³¼ ì¬í˜„ì„ ìœ„í•œ ì‹œë“œ ì„¤ì •

# ìƒí™©: í†µì¦ ìˆ˜ì¹˜(0~100)ê°€ ì–¼ë§ˆë‚˜ ì¤„ì–´ë“¤ì—ˆëŠ”ê°€? (ìˆ˜ì¹˜ê°€ ë†’ì„ìˆ˜ë¡ íš¨ê³¼ ì¢‹ìŒ)
# ê¸°ì¡´ ì§„í†µì œ(Aê·¸ë£¹): í‰ê·  50ë§Œí¼ ê°ì†Œ, í‘œì¤€í¸ì°¨ 10, í™˜ì 30ëª…
group_old = np.random.normal(loc=50, scale=10, size=30)

# ì‹ ê·œ ì§„í†µì œ(Bê·¸ë£¹): í‰ê·  55ë§Œí¼ ê°ì†Œ, í‘œì¤€í¸ì°¨ 10, í™˜ì 30ëª…
# (ìš°ë¦¬ëŠ” ì‹ ê·œ ì•½ì´ ë” ì¢‹ë‹¤ê³  ê°€ì •í•˜ê³  ë°ì´í„°ë¥¼ ë§Œë“¤ì—ˆìŠµë‹ˆë‹¤)
group_new = np.random.normal(loc=55, scale=10, size=30)

print(f"[ë°ì´í„° í™•ì¸]")
print(f"ê¸°ì¡´ ì§„í†µì œ íš¨ê³¼ í‰ê· : {group_old.mean():.2f}")
print(f"ì‹ ê·œ ì§„í†µì œ íš¨ê³¼ í‰ê· : {group_new.mean():.2f}")
print("-" * 30)


# ==========================================
# 2ë‹¨ê³„: ê°€ì„¤ ìˆ˜ë¦½ ë° ê²€ì • í†µê³„ëŸ‰ ì‚°ì¶œ
# ==========================================
# H0: New <= Old (íš¨ê³¼ê°€ ì—†ê±°ë‚˜ ê°™ë‹¤)
# H1: New > Old (ì‹ ê·œ ì•½ì´ ë” ìš°ìˆ˜í•˜ë‹¤) -> 'greater' ì˜µì…˜ ì‚¬ìš© (ë‹¨ì¸¡ ê²€ì •)

# ë…ë¦½ í‘œë³¸ T-ê²€ì • ìˆ˜í–‰
t_stat, p_value = stats.ttest_ind(
    group_new,       # ì‹ ê·œ ì•½ (ëŒ€ë¦½ê°€ì„¤ì˜ ì£¼ì¸ê³µ)
    group_old,       # ê¸°ì¡´ ì•½
    equal_var=False, # ë“±ë¶„ì‚° ê°€ì • ì•ˆ í•¨ (Welch's t-test, ì‹¤ë¬´ì ìœ¼ë¡œ ë” ì•ˆì „í•¨)
    alternative='greater' # ì¤‘ìš”: "ë” í¬ë‹¤(ìš°ìˆ˜í•˜ë‹¤)"ë¥¼ ê²€ì¦í•˜ë¯€ë¡œ ìš°ì¸¡ ê²€ì •
)

print(f"[ê²€ì • ê²°ê³¼]")
print(f"T-statistic (ì°¨ì´ì˜ í¬ê¸°): {t_stat:.4f}")
print(f"P-value (ìš°ì—°ì¼ í™•ë¥ ): {p_value:.4f}")
print("-" * 30)


# ==========================================
# 3ë‹¨ê³„: ìµœì¢… ê²°ì • (ë¹„ì¦ˆë‹ˆìŠ¤ ì˜ì‚¬ê²°ì •)
# ==========================================
alpha = 0.05  # ìœ ì˜ìˆ˜ì¤€ (ë¦¬ìŠ¤í¬ í—ˆìš© ë²”ìœ„ 5%)

print(f"[ìµœì¢… íŒë‹¨]")
if p_value < alpha:
    print(f"ê²°ê³¼: ê·€ë¬´ê°€ì„¤ ê¸°ê° (P-value {p_value:.4f} < {alpha})")
    print("í•´ì„: ì‹ ê·œ ì§„í†µì œì˜ íš¨ëŠ¥ì´ í†µê³„ì ìœ¼ë¡œ ìœ ì˜ë¯¸í•˜ê²Œ ìš°ìˆ˜í•©ë‹ˆë‹¤.")
    print("ğŸ‘‰ Action: ì‹ ê·œ ì•½ ì¶œì‹œ ìŠ¹ì¸ / ë§ˆì¼€íŒ… í¬ì¸íŠ¸ë¡œ í™œìš©")
else:
    print(f"ê²°ê³¼: ê·€ë¬´ê°€ì„¤ ì±„íƒ (P-value {p_value:.4f} >= {alpha})")
    print("í•´ì„: í‰ê· ì€ ì¡°ê¸ˆ ë†’ì•„ ë³´ì¼ì§€ ëª°ë¼ë„, í†µê³„ì ìœ¼ë¡œ ìš°ìˆ˜í•˜ë‹¤ê³  ë³¼ ìˆ˜ ì—†ìŠµë‹ˆë‹¤ (ìš°ì—°ì¼ ê°€ëŠ¥ì„± ë†’ìŒ).")
    print("ğŸ‘‰ Action: ì¶œì‹œ ë³´ë¥˜ / ì¶”ê°€ ì„ìƒ ì‹¤í—˜ í•„ìš”")
```

```python
#ì‚¬ë¡€: ì„œë¡œ ë‹¤ë¥¸ 3ê°€ì§€ ë¹„ë£Œ(A, B, C)ë¥¼ ì£¼ì—ˆì„ ë•Œ ì‹ë¬¼ì˜ í‚¤ ì„±ì¥ì— ì°¨ì´ê°€ ìˆëŠ”ê°€?
import pandas as pd
from scipy import stats

# ë°ì´í„° ì¤€ë¹„: ë¹„ë£Œ ì¢…ë¥˜ë³„ ì‹ë¬¼ í‚¤
group_a = [20, 22, 19, 21, 23]
group_b = [28, 30, 27, 29, 31]
group_c = [24, 25, 23, 26, 24]

# ì¼ì› ë¶„ì‚°ë¶„ì„ ì‹¤í–‰: scipyì˜ f_oneway í•¨ìˆ˜ ì‚¬ìš©
f_val, p_val = stats.f_oneway(group_a, group_b, group_c)

print(f"ì¼ì› ANOVA - F-í†µê³„ëŸ‰: {f_val:.4f}, P-value: {p_val:.4f}")
# í•´ì„: p_val < 0.05 ì´ë©´ ë¹„ë£Œë³„ ì„±ì¥ì— ìœ ì˜ë¯¸í•œ ì°¨ì´ê°€ ìˆìŒ
```

```python
### ANOVA F-ê²€ì • [ì´ì›ìœ í˜•]
```

```python
#ì‚¬ë¡€: 'ì„±ë³„'(ë‚¨/ì—¬)ê³¼ 'êµìœ¡ë°©ë²•'(ì˜¨ë¼ì¸/ì˜¤í”„ë¼ì¸)ì´ë¼ëŠ” ë‘ ê°€ì§€ ê¸°ì¤€ì— ë”°ë¼ ì‹œí—˜ ì ìˆ˜ê°€ ë‹¤ë¥¸ê°€?
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# ë°ì´í„° ì¤€ë¹„: ì„±ë³„, êµìœ¡ë²•, ì ìˆ˜
data = pd.DataFrame({
    'gender': ['M','M','M','F','F','F']*2,
    'method': ['On','On','On','On','On','On','Off','Off','Off','Off','Off','Off'],
    'score': [85, 88, 90, 92, 95, 91, 75, 78, 80, 82, 85, 79]
})

# ëª¨ë¸ ìˆ˜ì‹ ì„¤ì •: score ~ gender * method (ìƒí˜¸ì‘ìš© íš¨ê³¼ í¬í•¨)
model = ols('score ~ gender * method', data=data).fit()
anova_table = anova_lm(model)

print("ì´ì› ANOVA ê²°ê³¼:\n", anova_table)
```

```python
### ANOVA F-ê²€ì • [ë‹¤ì›ìœ í˜•]
```

```python
#ì‚¬ë¡€: 'ì§€ì—­', 'í•™ë ¥', 'ì§ì¢…' 3ê°€ì§€ ë³€ìˆ˜ê°€ 'ì—°ë´‰'ì— ì˜í–¥ì„ ì£¼ëŠ”ê°€? (ë³€ìˆ˜ê°€ 3ê°œ ì´ìƒ)
import pandas as pd
import numpy as np
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# 1. ê°€ìƒì˜ ì—°ë´‰ ë°ì´í„° ìƒì„± (df_salary ì •ì˜)
# ë¶„ì„ì„ ìœ„í•´ ê° ë³€ìˆ˜ë³„ë¡œ ì ì ˆí•œ ìƒ˜í”Œ ë°ì´í„°ë¥¼ ë§Œë“­ë‹ˆë‹¤.
data_size = 30
np.random.seed(42) # ê²°ê³¼ ì¬í˜„ì„ ìœ„í•´ ëœë¤ ì‹œë“œ ê³ ì •

df_salary = pd.DataFrame({
    'region': np.random.choice(['Seoul', 'Busan', 'Gwangju'], data_size),
    'education': np.random.choice(['Bachelor', 'Master', 'PhD'], data_size),
    'job_type': np.random.choice(['IT', 'Sales', 'HR'], data_size),
    'salary': np.random.randint(3000, 9000, data_size) # 3000~9000 ì‚¬ì´ì˜ ì—°ë´‰
})

# ë°ì´í„°ê°€ ì˜ ìƒì„±ë˜ì—ˆëŠ”ì§€ ìƒìœ„ 5ê°œë§Œ í™•ì¸
print("--- ìƒì„±ëœ ë°ì´í„° ì¼ë¶€ ---")
print(df_salary.head())
print("\n")

# 2. ë‹¤ì› ë¶„ì‚°ë¶„ì„ ëª¨ë¸ ì„¤ì •
# ìˆ˜ì‹ ì„¤ëª…: salary(ê²°ê³¼) ~ region + education + job_type(ì›ì¸ë“¤)
# ê° ë³€ìˆ˜ ì•ì— 'C()'ë¥¼ ë¶™ì—¬ ë²”ì£¼í˜•(ëª…ëª©í˜•)ì„ì„ ëª…ì‹œí•˜ë©´ ë” ì •í™•í•©ë‹ˆë‹¤.
formula = 'salary ~ C(region) + C(education) + C(job_type)'
model_multi = ols(formula, data=df_salary).fit()

# 3. ANOVA í…Œì´ë¸” ì¶œë ¥
anova_table = anova_lm(model_multi)

print("--- ë‹¤ì› ANOVA ë¶„ì„ ê²°ê³¼ ---")
print(anova_table)

# 4. ê²°ê³¼ í•´ì„ ê°€ì´ë“œ (P-value í™•ì¸)
# PR(>F) ì»¬ëŸ¼ì˜ ê°’ì´ 0.05ë³´ë‹¤ ì‘ì€ ë³€ìˆ˜ê°€ ìˆë‹¤ë©´, ê·¸ ë³€ìˆ˜ëŠ” ì—°ë´‰ì— ìœ ì˜ë¯¸í•œ ì˜í–¥ì„ ì£¼ëŠ” ê²ƒì…ë‹ˆë‹¤.
```

```python
### ANOVA F-ê²€ì • [ë°˜ë³µì¸¡ì • ìœ í˜•]
```

```python
#ì‚¬ë¡€: ë™ì¼í•œ í™˜ì 5ëª…ì—ê²Œ ì•½ì„ íˆ¬ì—¬í•œ í›„ 'ì§ì „', '1ì£¼ í›„', '2ì£¼ í›„'ì˜ í˜ˆì•• ë³€í™”ë¥¼ ì¸¡ì •í•¨.
from statsmodels.stats.anova import AnovaRM

# ë°ì´í„° ì¤€ë¹„: ë™ì¼ ì¸ë¬¼(id)ì— ëŒ€í•´ ë°˜ë³µ ì¸¡ì •(time)
df_rm = pd.DataFrame({
    'id': [1, 2, 3, 4, 5] * 3,
    'time': ['pre', 'pre', 'pre', 'pre', 'pre', '1w', '1w', '1w', '1w', '1w', '2w', '2w', '2w', '2w', '2w'],
    'bp': [150, 145, 155, 160, 148, 140, 138, 145, 150, 142, 130, 125, 135, 140, 132]
})

# ë°˜ë³µì¸¡ì • ANOVA ì‹¤í–‰: ëŒ€ìƒ(id), ì¢…ì†ë³€ìˆ˜(bp), ë…ë¦½ë³€ìˆ˜(time) ì§€ì •
res = AnovaRM(data=df_rm, depvar='bp', subject='id', within=['time']).fit()
print("ë°˜ë³µì¸¡ì • ANOVA ê²°ê³¼:\n", res)
```

```python
#ì‚¬ë¡€: 'ì‹ ì•½ íˆ¬ì—¬ ì—¬ë¶€'ì— ë”°ë¼ 'í˜ˆì••'ê³¼ 'ì½œë ˆìŠ¤í…Œë¡¤ ìˆ˜ì¹˜'(ê²°ê³¼ê°’ì´ 2ê°œ ì´ìƒ)ê°€ ë™ì‹œì— ë³€í•˜ëŠ”ê°€?
from statsmodels.multivariate.manova import MANOVA

# ë°ì´í„° ì¤€ë¹„: ê²°ê³¼ê°’ì´ ë‘ ê°œ(blood_pressure, cholesterol)
df_manova = pd.DataFrame({
    'group': ['A', 'A', 'A', 'B', 'B', 'B'],
    'blood_p': [120, 125, 122, 140, 145, 142],
    'cholesterol': [200, 210, 205, 240, 250, 245]
})

# MANOVA ì‹¤í–‰: ì™¼ìª½ ë³€ìˆ˜ë“¤ì— ë¦¬ìŠ¤íŠ¸ í˜•íƒœë¡œ ê²°ê³¼ê°’ë“¤ì„ ë„£ìŒ
ma_model = MANOVA.from_formula('blood_p + cholesterol ~ group', data=df_manova)
print("MANOVA ê²°ê³¼:\n", ma_model.mv_test())
```
