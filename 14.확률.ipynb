```python
exp_p = 0.5 #확률
trial_count = 10 #시행
head_count = 0 #앞면이 나온 건수
#동전의 앞면이 나올 확률 50%
#10번 시행했을 때 상대도수와 확률 비교


for cnt in range(trial_count): 
    if random.random() < exp_p :
        head_count += 1
        
print(f'앞면이 나온 횟수 : {head_count}')
actual_freq = head_count/trial_count

print(f'예상확률 : {exp_p*100}%')
print(f'실제결과(상대도수) : {actual_freq*100}%')
```

```python
import math
total_balls = 7
white_balls = 4
black_balls = 3
pick_count = 3 #white2, black1

#전체 경우의 수
total_case = math.comb(total_balls, pick_count)
#흰공이 2개 마올 경우의 수(4,2)
white_case = math.comb(white_balls, 2)
#검은공 1개 내올 경우의 수(3,1)
black_case = math.comb(black_balls, 1)

#흰공2, 검은공1 경우의 수
event_case = white_case * black_case

prob = event_case / total_case
print(f'계산한 확률 : {prob*100:.2f}')


```

```python
#시뮬레이션
def run_simulation(trials):
    act_freq = 0
    success_cnt = 0
    sample_space = ['W','W','W','W','B','B','B'] #표본공간
    for i in range(trials):
        drawn_balls = random.sample(sample_space, 3)
        if drawn_balls.count('W') == 2 and drawn_balls.count('B') == 1 :
            success_cnt += 1 #흰공2, 검공1개 나온 횟수
    act_freq = success_cnt / trials #상대도수
    return act_freq
```

```python
trials = 10
simulation_prob = run_simulation(trials)
print(f'시뮬레이션횟수 {trials}, 결과확률 {simulation_prob:.2f}')
```

```python
# 종속사건
total = 10
win = 2
prob_A = win / total #A가 당첨될 확률
prob_B = (win-1) / (total -1) #A 다음에 B가 당첨될 확률 P(B/A)
prob_AB = prob_A * prob_B #A확률  #B확률의 곱 = P(A) * P(B/A)
print(f'종속사건 AB의 확률 : {prob_AB:.2f}')
```

```python
# 전체 표본 공간
sample_space = [1, 2, 3, 4, 5, 6]
```

```python
# 사건 A: 짝수
event_A = [x for x in sample_space if x % 2 == 0]
```

```python
# 사건 B: 4의 약수
event_B = [x for x in sample_space if 4% x == 0]
```

```python
#교집합 A ∩ B
event_A_and_B = list(set(event_A) & set(event_B))
```

```python
# 확률 계산
P_A = len(event_A) / len(sample_space)
P_A_and_B = len(event_A_and_B) / len(sample_space)
```

```python
# 조건부확률 P(B | A)
P_B_given_A = P_A_and_B / P_A

P_B_given_A
```

# 기대값 구하기 [확률에서는 평균을 기대값으로 구분한다]

```python
x_values = list( range(1,4) )
p_X = [1/6, 3/6, 2/6]
exp_value = 0
for i in range( len(x_values) ):
    exp_value += (x_values[i]*p_X[i])
print(f'기대값 : {exp_value:.5f}')
```
