```python
import random

# 1. 선택지 정의
choices = ["가위", "바위", "보"]

# 2. 플레이어 입력
player = input("가위, 바위, 보 중 하나를 입력하세요: ")

# 입력값 검증
if player not in choices:
    print("잘못된 입력입니다. 가위, 바위, 보 중에서 선택하세요.")
else:
    # 3. 컴퓨터 랜덤 선택
    computer = random.choice(choices)

    print(f"플레이어 선택: {player}")
    print(f"컴퓨터 선택: {computer}")

    # 4. 승패 판정
    if player == computer:
        result = "무승부"
    elif (
        (player == "가위" and computer == "보") or
        (player == "바위" and computer == "가위") or
        (player == "보" and computer == "바위")
    ):
        result = "승리"
    else:
        result = "패배"

    # 5. 결과 출력
    print(f"결과: {result}")
```

```python
def check_access(user_role):
    if user_role == "admin":
        return "관리자 페이지에 접근 가능합니다."
    elif user_role == "user":
        return "일반 사용자 페이지에 접근 가능합니다."
    else:
        return "접근 권한이 없습니다."
#테스트
print(check_access("admin")) #관리자 페이지에 접근 가능합니다.
print(check_access("user")) #일반 사용자 페이지에 접근 가능합니다.
print(check_access("guest")) #접근 권한이 없습니다.
```
